```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import matplotlib.gridspec as gridspec
```

```python
csv_file = 'data\\_788_h.csv'
dateparse = lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
df = pd.read_csv(csv_file, parse_dates = ['Datetime'], date_parser=dateparse, index_col = ['Datetime'])

start_date = '2023-01-01'
start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d")
    
end_date = datetime.datetime.now()
mask = (( df.index  >= start_date ) & ( df.index  <= end_date ) & (df['Volume'] > 15000000))
df = df[mask]

print(df.head())
```

```python
# __first plot__ ???
ticker = '788.hk'
fig = plt.figure(figsize=(12,4))
fig.suptitle('Volume ' + str(ticker))
plt.bar(df.index.strftime('%Y-%m-%d %H:%M'), df['Volume'], label='Volume',color='black', alpha=0.5, width = 0.02)
plt.xlabel('DateTime')
plt.xticks(rotation=90, ha='right')
plt.show()
```
